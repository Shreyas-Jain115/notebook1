```python
df1 = pd.DataFrame({
    "Key": ["K1", "K2", "K3", "K4"],
    "A":   [10, 20, 30, 40],
    "B":   [11, 21, 31, 41]
})

# Second DataFrame with Key, C, D
df2 = pd.DataFrame({
    "Key": ["K1", "K3", "K2", "K4"],
    "C":   [50, 60, 70, 80],
    "D":   [51, 61, 71, 81]
})
```

```python
df1
```

```python
pd.merge(df1,df2,how='inner',on='Key') #how='inner' is default
```

```python
left = pd.DataFrame({
    "Key1": ["K0", "K0", "K1", "K2"],
    "Key2": ["K0", "K1", "K0", "K1"],
    "A":   [10, 20, 30, 40],
    "B":   [11, 21, 31, 41]
})

# Second DataFrame with Key, C, D
right = pd.DataFrame({
    "Key1": ["K0", "K1", "K1", "K2"],
    "Key2": ["K0", "K0", "K0", "K0"],
    "C":   [50, 60, 70, 80],
    "D":   [51, 61, 71, 81]
})
```

```python
pd.merge(left,right,on=['Key1','Key2'])
```

```python
pd.merge(left,right,on=['Key1','Key2'],how='outer')
```

```python
pd.merge(left,right,on=['Key1','Key2'],how='left')
```

```python
pd.merge(left,right,on=['Key1','Key2'],how='right')
```
